```python
import pandas as pd

# Load the dataset
dataset_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(dataset_url)

# Display the first few rows of the dataset (optional, just for context)
print("First 5 rows of the dataset:")
print(df.head(), "\n")

# Display summary of dataset: data types, non-null counts
print("Dataset Info (Data types and non-null counts):")
print(df.info(), "\n")

# Count of missing values per column
print("Missing Values per Column:")
print(df.isnull().sum(), "\n")

# Data types of each column
print("Column Data Types:")
print(df.dtypes)
```

```python
# Basic statistics for numerical columns
print("Basic Statistics for Numerical Columns:")
print(df.describe())
```

```python
# Unique values in the 'Country' column
print("Unique values in 'Country' column:")
print(df['Country'].unique(), "\n")

# Count of each unique country (to detect rare or inconsistent values)
print("Value counts in 'Country' column:")
print(df['Country'].value_counts(dropna=False), "\n")

# Check for leading/trailing spaces or inconsistent capitalization
print("Normalized unique 'Country' values:")
print(df['Country'].str.strip().str.title().value_counts())
```

```python
# Standardize 'Country' column
df['Country'] = df['Country'].str.strip().str.title()

country_mapping = {
    'Usa': 'USA',
    'United States': 'USA',
    'Uk': 'UK',
    'United Kingdom': 'UK',
    'England': 'UK',
    'Republic Of India': 'India',
    'India ': 'India',
    'Na': None,
    'None': None
}

df['Country'] = df['Country'].replace(country_mapping)

# Standardize 'EdLevel' column
df['EdLevel'] = df['EdLevel'].str.strip().str.title()

edlevel_mapping = {
    "Bachelor’s Degree": "Bachelor’s degree",
    "Master’s Degree": "Master’s degree",
    "Some College": "Some college",
    "High School": "High school",
    "High School Diploma": "High school",
    "Associate Degree": "Associate degree",
    "Na": None,
    "None": None
}

df['EdLevel'] = df['EdLevel'].replace(edlevel_mapping)

# Optional: Check unique values after cleaning
print("Unique values in 'Country':")
print(df['Country'].unique(), "\n")

print("Unique values in 'EdLevel':")
print(df['EdLevel'].unique())
```

```python
# One-hot encode the 'Employment' column
employment_encoded = pd.get_dummies(df['Employment'], prefix='Employment')

# Concatenate the encoded columns back to the original DataFrame
df = pd.concat([df, employment_encoded], axis=1)

# Optional: Drop the original 'Employment' column if no longer needed
df.drop('Employment', axis=1, inplace=True)

# Preview the DataFrame
print("DataFrame with One-Hot Encoded 'Employment':")
print(df.head())
```

```python
# Count missing values in each column
missing_values = df.isnull().sum()

# Filter columns with at least 1 missing value and sort by count descending
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

# Display the result
print("Columns with the highest number of missing values:")
print(missing_values)
```

```python
# Impute missing values in ConvertedCompYearly with the median
median_value = df['ConvertedCompYearly'].median()
df['ConvertedCompYearly'].fillna(median_value, inplace=True)

print("Missing values in ConvertedCompYearly after median imputation:", df['ConvertedCompYearly'].isnull().sum())
```

```python
# Impute missing values in ConvertedCompYearly with the mean
mean_value = df['ConvertedCompYearly'].mean()
df['ConvertedCompYearly'].fillna(mean_value, inplace=True)

print("Missing values in ConvertedCompYearly after mean imputation:", df['ConvertedCompYearly'].isnull().sum())
```

```python
# Find the most frequent (mode) value in the column
most_frequent = df['RemoteWork'].mode()[0]

# Fill missing values with the mode
df['RemoteWork'].fillna(most_frequent, inplace=True)

# Verify
print("Missing values in RemoteWork after imputation:", df['RemoteWork'].isnull().sum())
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize scaler
scaler = MinMaxScaler()

# Reshape is required since scaler expects 2D input
df['ConvertedCompYearly_Scaled'] = scaler.fit_transform(df[['ConvertedCompYearly']])

# Preview the result
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Scaled']].head())
```

```python
import numpy as np

# Add a small constant to avoid log(0) error
df['ConvertedCompYearly_Log'] = np.log1p(df['ConvertedCompYearly'])

# Preview original vs. log-transformed values
print(df[['ConvertedCompYearly', 'ConvertedCompYearly_Log']].head())
```

```python
# Clean 'YearsCodePro' by converting strings to numeric values
df['YearsCodePro'] = df['YearsCodePro'].replace({
    'Less than 1 year': 0.5,
    'More than 50 years': 51
}).astype(float)

# Define experience levels
def classify_experience(years):
    if pd.isnull(years):
        return None
    elif years < 3:
        return 'Beginner'
    elif 3 <= years < 7:
        return 'Intermediate'
    else:
        return 'Expert'

# Apply the function to create new column
df['ExperienceLevel'] = df['YearsCodePro'].apply(classify_experience)

# Preview the result
print(df[['YearsCodePro', 'ExperienceLevel']].head())
```
